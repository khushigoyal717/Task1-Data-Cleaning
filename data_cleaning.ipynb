```
#Load the Dataset and View Basic Info
import pandas as pd

df = pd.read_csv("marketing.csv", sep='\t')
print("Initial shape:", df.shape)
df.head()
```

```
#check missing values
df.isnull().sum()
```

```
#handling missing values
df['Income'].fillna(df['Income'].mean(), inplace=True)
```

```
#Remove Duplicate Rows
df = df.drop_duplicates()
```

```
#Standardize Text Columns
df['Education'] = df['Education'].str.strip().str.lower()
df['Marital_Status'] = df['Marital_Status'].str.strip().str.lower()
```

```
#Convert Date Column to Proper Format
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['Dt_Customer'] = df['Dt_Customer'].dt.strftime('%Y-%m-%d')
```

```
#Clean Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```
#Ensure Data Types Are Correct
df['income'] = df['income'].astype(float)
```

```
# Save cleaned data to a new CSV
df.to_csv("cleaned_customer_personality.csv", index=False)
print("Cleaned dataset saved as cleaned_customer_personality.csv")
```
